# Task 3 : Dictonary

```python
# 1. price dictionary with product name as key and price as value
price_dict = {
    "Laptop": 51000,
    "Smartphone": 20000.50,
    "Headphones": 3000,
    "Camera": 25000,
    "Smartwatch": 15000.75,
    "Tablet": 22000
}
type(price_dict)
```

```python
print("price dictionary:", price_dict)
```

```python
# 2. perform dictionary operations like update price, add new product with price, delete a product price
# Add new product with price
price_dict["Printer"] = 12000
# Update price of an existing product
price_dict["Laptop"] = 49999.99
# Delete a product price
del price_dict["Smartwatch"]
print("Updated price dictionary:", price_dict)
```

```python
# 3. print the average price of all products in the dictionary
average_price = sum(price_dict.values()) / len(price_dict)
print("Average price of products:", average_price)
```

```python
# Optional : print product with max and min price
max_price_product = max(price_dict, key=price_dict.get)
min_price_product = min(price_dict, key=price_dict.get)
print("Product with max price:", max_price_product, "Price:", price_dict[max_price_product])
print("Product with min price:", min_price_product, "Price:", price_dict[min_price_product])
```

# Task 4: Combined

```python
# 1. create a list of tuples using products and price dictionary
products_category = ["Smartphone", "Headphones", "Camera", "Tablet"]
price_dictionary = {
    "Apple iPhone": 20000.50,
    "Sony Headphones": 3000,
    "Fujitsu Camera": 25000,
    "Samsung Tablet": 22000
}
catalog = list(zip(price_dictionary.keys(), price_dictionary.values(), products_category))
print("List of catalog tuples:", catalog)
```

```python
# 2. From product catalog list of tuples, create a dictionary with category_to_product mapping each category to list of product names in the category
category_to_product = {}
for product, price, category in catalog:
    if category not in category_to_product:
        category_to_product[category] = []
    category_to_product[category].append(product)
print("Category to product mapping:", category_to_product)
```

```python
#add new entry in catalog
new_product = ("OnePlus Nord", 35000, "Smartphone")
catalog.append(new_product)
print("Updated catalog:", catalog)
```

```python
# 3. print all products values in catalog  which has maximum number of  products in the category
category_count = {}
for product, price, category in catalog:
    if category not in category_count:
        category_count[category] = 0
    category_count[category] += 1
max_category = max(category_count, key=category_count.get)
print("Category with maximum products:", max_category)
```
