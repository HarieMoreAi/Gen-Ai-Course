## Task1 : Discount Rules (if/elif/else)

```python
#1. read integer order_amount from user input input()
order_amount = int(input("Enter the order amount"))
print("Order amount is", order_amount)
```

```python
#2. discount processing
if order_amount >= 2000:
    print("Discount is 15%")
    order_amount = order_amount - (order_amount * 15 / 100)
elif order_amount >= 1500 and order_amount < 2000:
    print("Discount is 10%")        
    order_amount = order_amount - (order_amount * 10 / 100)
elif order_amount >= 1000 and order_amount < 1500:
    print("Discount is 7%")
    order_amount = order_amount - (order_amount * 7 / 100)  
else:
    print("No discount")
print("Final order amount is", order_amount)
```

```python
#3. convert input to number and handle case when input is not a number
try:
    product_amount = int(input("Enter the order amount"))
    print("Order amount is", product_amount)
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

```python
#Optional: add tax 5% to order amount and print subtotal, tax amount and total amount
tax_rate = 5
tax_amount = order_amount * tax_rate / 100
subtotal = order_amount
total_amount = subtotal + tax_amount
print("Subtotal:", subtotal)
print("Tax amount:", tax_amount)
print("Total amount:", total_amount)
```

## Task2 : Process Multiple Orders (for loop)

```python
#1. apply discount on order amount and print summary table : order_amount -> dicount % -> final_amount
orders = [1200, 2500, 800, 1750, 3000]
print("Order Amount | Discount % | Final Amount")
for order in orders:
    if order >= 2000:
        discount = 15
    elif order >= 1500 and order < 2000:
        discount = 10        
    elif order >= 1000 and order < 1500:
        discount = 7
    else:
        discount = 0
    final_amount = order - (order * discount / 100)
    print(f"{order:12} | {discount:10} | {final_amount:10}") #applied spacing for better alignment
```

```python
#2. compute and print total revenue from all orders after discount
total_revenue = 0
for order in orders:
    if order >= 2000:
        discount = 15
    elif order >= 1500 and order < 2000:
        discount = 10        
    elif order >= 1000 and order < 1500:
        discount = 7
    else:
        discount = 0
    final_amount = order - (order * discount / 100)
    total_revenue += final_amount

print(f"Total Revenue: {total_revenue}")
```

```python
#Optional: print number of orders that received discount > 0%
discounted_orders_count = 0
for order in orders:
    if order >= 2000:
        discount = 15
    elif order >= 1500 and order < 2000:
        discount = 10        
    elif order >= 1000 and order < 1500:
        discount = 7
    else:
        discount = 0
    if discount > 0:
        discounted_orders_count += 1

print(f"Number of orders that received discount > 0%: {discounted_orders_count}")
```

## Task3 : User Menu (while loop + break/continue)

```python
#1. while-loop menu until user enters q to quit. Menu options: 1. add order, 2. view all orders & totals after applying discount, q. quit
orders = []
while True:
    print("Menu:")
    print("1. Add order")
    print("2. View all orders & totals after applying discount")
    print("q. Quit")
    choice = input("Enter your choice: ")
    
    if choice == '1':
        order_amount = int(input("Enter the order amount: "))
        orders.append(order_amount)
        print("Order added successfully.")
    
    elif choice == '2':
        if not orders:
            print("No orders to display.")
        else:
            print("Order Amount | Discount % | Final Amount")
            total_revenue = 0
            for order in orders:
                if order >= 2000:
                    discount = 15
                elif order >= 1500 and order < 2000:
                    discount = 10        
                elif order >= 1000 and order < 1500:
                    discount = 7
                else:
                    discount = 0
                final_amount = order - (order * discount / 100)
                total_revenue += final_amount
                print(f"{order:12} | {discount:10} | {final_amount:10}")
            print(f"Total Revenue: {total_revenue}")
    
    elif choice == 'q':
        print("Exiting the program. Goodbye!")
        break
    
    else:
        print("Invalid choice. Please select a valid option from the menu.")
```

## Task4 : Loop controls with conditions (break & continue)

```python
#1. use of break, skip, and adding amount to total sales based on type of sales value
daily = [200, 150, 0, 400, 50, -1, 300]
total_sales = 0
for sale in daily:
    if sale < 0:
        print("Negative sales value encountered. Corrupted data. Stopping processing.")
        break
    elif sale == 0:
        print("Zero sales value encountered. Skipping this entry.")
        continue
    total_sales += sale
    print(f"Added {sale} to total sales. Current total: {total_sales}")
```

```python
#2. print the final total
print(f"Final Total Sales: {total_sales}")
```
